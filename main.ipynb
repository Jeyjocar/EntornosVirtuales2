```python

personas = ['Juan','Julian','Pupo','Lisbeth','Jeyfrey']
edades = [25, 20, 35, 29, 40]

#* lienzos

notas = np.random.normal(loc=45, scale=10, size=(10000))
indice_alumnos = [x for x in range(1,10001)]

frecuencia_notas = np.array([],dtype='int32')

nota_redondeada = np.array([])

# implementamos un bucle for 

for i in notas:
    z = round(i)
    nota_redondeada = np.append(nota_redondeada, z)

for j in range(-30,131):
    frecuencia_notas = np.append(frecuencia_notas, np.count_nonzero(nota_redondeada == j))
    
notas_posibles = np.array([x for x in range(-30,131)])

#creamos diccionario
datos = pd.DataFrame({'Posibles Notas':  notas_posibles, 'Notas frecuentes': frecuencia_notas})

bar_alumnos = pd.DataFrame({'indice' : indice_alumnos, 
                            'alumnos' : indice_alumnos, 
                            'notas_Promedio' : notas
                            })

bar_alumnos.set_index('indice', inplace=True)

#? formateo de graficas(2x2)

fig,puntero = plt.subplots(nrows=2, ncols=2, figsize=(12,10))


#! gráfico N° 1

puntero[0][0].bar(personas, edades, color='#2D2EEA')
puntero[0][0].set_title('edad de los participantes')
puntero[0][0].set_xlabel('personas')
puntero[0][0].set_ylabel('edades')

#* gráfico N° 2

draw_1 = ECDF(x=edades)
puntero[0][1].plot(personas, edades, color='c', marker='d', linestyle='--')
puntero[0][1].set_title('curva de edades')
puntero[0][1].set_xlabel('edad por personas')
puntero[0][1].set_ylabel('Fa')

#? gráfico N° 3

histo,bins,pad = puntero[1][0].hist(x=bar_alumnos,bins=50) 
puntero[1][0].set_title('promedio de notas')
puntero[1][0].set_xlabel('ranking')
puntero[1][0].set_ylabel('N° alumnos')

#! gráfico N° 4

sigma=10

mid=45

# y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma(bins-mid)**2)))

puntero[1][0].set_title('promedio de notas')
puntero[1][0].set_xlabel('ranking')
puntero[1][0].set_ylabel('N° alumnos')

puntero[1][1].plot(notas_posibles, frecuencia_notas, color='r')
puntero[1][1].set_title('resultado de ejercicio')
puntero[1][1].set_xlabel('ranking')
puntero[1][1].set_ylabel('N° alumnos')
```
